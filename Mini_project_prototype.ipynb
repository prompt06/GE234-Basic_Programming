```python
import folium
from folium.plugins import MousePosition
import pandas as pd

# --- 1. ตัด Logic ให้เหลือแต่เนื้อๆ (Pre-calculated) ---
def quick_analyze(lat, lon):
    # วิเคราะห์ตามโซนภูมิศาสตร์จริง
    if 13.9 <= lat <= 14.2:
        res = {"prov": "ปทุมธานี", "agri": "ข้าว/บัว", "type": "ที่ราบลุ่มน้ำท่วมถึง"}
    elif 13.4 <= lat < 13.9:
        res = {"prov": "ราชบุรี/กาญจนบุรี", "agri": "สับปะรด/อ้อย", "type": "ที่ดอนทนแล้ง"}
    else:
        res = {"prov": "พื้นที่ทั่วไป", "agri": "พืชไร่", "type": "วิเคราะห์ตามชุดดิน"}
    return res

# --- 2. ฟังก์ชันแสดงผลแบบ Static (ชัวร์ 100% ตอนพรี) ---
def show_map_with_result(lat, lon):
    data = quick_analyze(lat, lon)

    # สร้างแผนที่
    m = folium.Map(location=[lat, lon], zoom_start=12, tiles='CartoDB positron')

    # ปักหมุดตรงที่จิ้ม
    popup_text = f"<b>{data['prov']}</b><br>เกษตร: {data['agri']}<br>ลักษณะ: {data['type']}"
    folium.Marker([lat, lon], popup=popup_text, icon=folium.Icon(color='red', icon='info-sign')).add_to(m)

    # กล่องสรุปผลด้านข้าง (แสดงผลผ่าน Markdown ของ Colab)
    print(f"📍 ผลการวิเคราะห์พิกัด: {lat}, {lon}")
    print(f"🏢 จังหวัด: {data['prov']}")
    print(f"🌾 ศักยภาพเกษตร: {data['agri']}")
    print(f"⛰️ ลักษณะทางกายภาพ: {data['type']}")

    return m

# --- 3. วิธีพรีเซนต์ (Manual Click & Run) ---
# ให้เพื่อนก๊อปพิกัดจากแผนที่มาใส่ตรงนี้แล้วกดรัน จะได้แผนที่พร้อมหมุดและคำวิเคราะห์ทันที
test_lat, test_lon = 13.58, 99.83 # ลองเปลี่ยนเลขตรงนี้ (ราชบุรี)
show_map_with_result(test_lat, test_lon)
```
